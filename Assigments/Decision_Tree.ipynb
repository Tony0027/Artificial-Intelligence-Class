```python
# Load libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt

# Load data
cancer_data = load_breast_cancer()
data = pd.DataFrame(cancer_data['data'], columns=cancer_data['feature_names'])
labels = pd.Series(cancer_data['target'])

# Data preprocessing (80% training, 20% testing split)
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, test_size=0.2, random_state=27)

# Model selection and training
model = DecisionTreeClassifier(random_state=27)
model.fit(train_data, train_labels)

# Model testing
predictions = model.predict(test_data)

# Calculate accuracy and classification report
accuracy = accuracy_score(test_labels, predictions)
report = classification_report(test_labels, predictions)

# Display accuracy and classification report
print(f"Accuracy: {accuracy:.4f}")
print("Classification Report:")
print(report)

# Visualize the decision tree
plt.figure(figsize=(30,15), dpi=300)
plot_tree(model, feature_names=cancer_data['feature_names'], class_names=cancer_data['target_names'], filled=True)
plt.show()
```
